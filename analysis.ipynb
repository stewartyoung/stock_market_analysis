```python
# Add the stocker tool
from stocker import Stocker
import matplotlib as plt
```

```python
# Initialise an instance of the stocker class using Microsoft as an example
microsoft = Stocker('MSFT')
```

```python
# stock is an attribute of the microsoft object
stock_history = microsoft.stock
# calling the method head on the stock history displays the data
stock_history.head()
```

```python
# We can plot the ENTIRE HISTORY of the stock by calling the plot_stock method
microsoft.plot_stock()
```

```python
# We can plot the stock over A SPECIFIED RANGE OF TIME by calling the plot_stock method with
# extra parameters
microsoft.plot_stock(start_date = '2000-01-03',  end_date = '2018-01-16',  
                     stats = ['Daily Change', 'Adj. Volume'],  plot_type='pct')
```

```python
microsoft.buy_and_hold(start_date='1986-03-13', 
                       end_date='2018-01-16', nshares=100)
```

```python
microsoft.buy_and_hold(start_date='1999-01-05', 
                      end_date='2002-01-03', nshares=100)
```

```python
model, model_data = microsoft.create_prophet_model()
```

```python
# model and model_data are objects from the previous cell
model.plot_components(model_data)
plt.pyplot.show()
```

```python
# If we believe there might be a weekly pattern, we can add that 
# in to the prophet model by changing the weekly_seasonality attribute of the Stocker object
print(microsoft.weekly_seasonality)
microsoft.weekly_seasonality = True
print(microsoft.weekly_seasonality)
```
